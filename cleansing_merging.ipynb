```python
# List of months
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# List of years to process
years = ['2023', '2024']

# Initialize an empty list to hold the DataFrames for each year and month
all_dfs = []

# Loop through each year and month
for year in years:
    for month in months:
        # Construct the file path for the current month and year
        input_file = f'data/{month}_{year}.csv'
        
        # Check if the file exists before attempting to read
        if os.path.exists(input_file):
            # Read the CSV file, skip the first two rows, and set the third row as the header
            df = pd.read_csv(input_file, skiprows=2, encoding='ISO-8859-1')
            
            # Drop the last two rows
            df = df.iloc[:-2].reset_index(drop=True)
            
            # Check if the current file is dec_2024 to apply the 24-hour format for date
            if month == 'dec' and year == '2024':
                # Convert Date column to datetime using the 24-hour format
                df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y %H:%M').dt.date
            else:
                # Convert Date column to datetime using the 12-hour AM/PM format
                df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M %p').dt.date
            
            # Append the cleansed DataFrame to the list of DataFrames
            all_dfs.append(df)

# Concatenate all DataFrames in the list to form one DataFrame for both years
df_full = pd.concat(all_dfs, ignore_index=True)

# Save the combined DataFrame to a CSV file for the full dataset
output_file = 'data/final_data/whole_sales.csv'
df_full.to_csv(output_file, index=False)
```
